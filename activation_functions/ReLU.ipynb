## ReLU : rectified linear units (activation function 1)


used as a valid activation function because during backpropagation when we apply chain rule, we either get 0,1

relu is applied to every element 

i.e. 1 is returned if x>0 else 0 returned. 

relu and softmax are often used as pairs,

relu for the hidden layers and softmax for the final layer. 


[ReLU blog on medium](https://medium.com/@meetkp/understanding-the-rectified-linear-unit-relu-a-key-activation-function-in-neural-networks-28108fba8f07)


```python
import matplotlib.pyplot as plt
import numpy as np
data = np.linspace(-10,10)
def relu(x):
    return np.maximum(x,0)
res = relu(data)
plt.plot(res)
```

## Leaky ReLU

```python

def relu(x):
    return np.maximum(x,0.1*x)
res = relu(data)
plt.plot(res)
```
